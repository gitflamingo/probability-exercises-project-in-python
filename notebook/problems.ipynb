## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

# Method 1 

# Initialize variables
conteoResultados = 0
total_simulaciones = 1000

# Simulation of dice rolls and probability calculation
for i in range(total_simulaciones):
    dado1 = np.random.randint(1, 7)  # Simulate roll of the first die
    dado2 = np.random.randint(1, 7)  # Simulate roll of the second die
    puntuacionTotal = dado1 + dado2  # Calculate the total score

    # Check if puntuacionTotal is even or greater than 7
    if puntuacionTotal % 2 == 0 or puntuacionTotal > 7:
        conteoResultados += 1

# Calculate the estimated probability
probabilidadEstimada = conteoResultados / total_simulaciones

print("The estimated probability is:", probabilidadEstimada)

```

```python
import numpy as np

totalSimulations = 1000

# Simulating the roll of two dice #
simulations = np.random.randint(1, 7, size=(totalSimulations, 2))

# Calculate the total score for each roll (sum of the two dice)
totalScore = np.sum(simulations, axis=1)

# Identify the rolls that meet the condition (even sum or greater than 7)
desiredResults = (totalScore % 2 == 0) | (totalScore > 7)

# Calculate the estimated probability
estimatedProbability = np.mean(desiredResults)
print(estimatedProbability)

```

```python
import random

# Create the box of balls
ball_box = {}
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# Perform the experiment and calculate probabilities simultaneously
totalExperiments = 1000
countColors = {"White": 0, "Red": 0, "Green": 0}

for i in range(totalExperiments):
    random_ball = random.choice(list(ball_box.values())) #The line of code random_ball = random.choice(list(ball_box.values())) is selecting a ball at random from the box of balls (ball_box)
    countColors[random_ball] += 1 #Increment that adds 1 to the current count of the selected color.

# Calculate probabilities
totalExperimentsFloat = float(totalExperiments)
probabilityWhite = countColors["White"] / totalExperimentsFloat
probabilityRed = countColors["Red"] / totalExperimentsFloat
probabilityGreen = countColors["Green"] / totalExperimentsFloat

# Print probabilities
print("Probability of selecting a white ball:", probabilityWhite)
print("Probability of selecting a red ball:", probabilityRed)
print("Probability of selecting a green ball:", probabilityGreen)  
```
