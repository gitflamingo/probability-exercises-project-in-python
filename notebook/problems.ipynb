## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
#Method 1 

# Inicializar variables
conteoResultados = 0
total_simulaciones = 1000

# Simulación de lanzamiento de dados y cálculo de probabilidad
for i in range(total_simulaciones):
    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)
    puntuacionTotal = dado1 + dado2

    # Verificar si la puntuacionTotal es par o mayor que 7
    if puntuacionTotal % 2 == 0 or puntuacionTotal > 7:
        conteoResultados += 1

# Cálculo de la probabilidad
probabilidadEstimada = conteoResultados / total_simulaciones

print("La probabilidad estimada es:", probabilidadEstimada)
```

```python
import numpy as np

totalSimulaciones = 1000

# Simulación del lanzamiento de dos dados #
simulaciones = np.random.randint(1, 7, size=(totalSimulaciones, 2))

# Calcula la puntuación total para cada lanzamiento (suma de los 2 dados)
puntuacionTotal = np.sum(simulaciones, axis=1)

# Encuentra los lanzamientos que cumplen con la condición (suma par o mayor que 7)
resultadosDeseados = (puntuacionTotal % 2 == 0) | (puntuacionTotal > 7)

# Calcula la probabilidad estimada
probabilidadEstimada = np.mean(resultadosDeseados)
print(probabilidadEstimada)
```

```python
import random

# Crear la caja de bolas
ball_box = {}
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# Realizar experimento y calcular las probabilidades al mismo tiempo
totalExperiments = 1000
countColors = {"White": 0, "Red": 0, "Green": 0}

for i in range(totalExperiments):
    random_ball = random.choice(list(ball_box.values())) #La línea de código random_ball = random.choice(list(ball_box.values())) está seleccionando una bola al azar de la caja de bolas (ball_box)
    countColors[random_ball] += 1 #Incremento que suma 1 al valor actual del contador del color.

# Calcular las probabilidades
totalExperimentsFloat = float(totalExperiments)
probabilidadWhite = countColors["White"] / totalExperimentsFloat
probabilidadRed = countColors["Red"] / totalExperimentsFloat
probabilidadGreen = countColors["Green"] / totalExperimentsFloat

# Imprimir probabilidades
print("Probabilidad de seleccionar una bola blanca:", probabilidadWhite)
print("Probabilidad de seleccionar una bola roja:", probabilidadRed)
print("Probabilidad de seleccionar una bola verde:", probabilidadGreen)


     
```
